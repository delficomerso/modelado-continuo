## 1.1

Defino la función sigmoidea con los valores por default $\rho=0$ y $c=1$ para obtener $S(x)=\frac{1}{1+e^{-x}}$

```python
def sigmoidea(x, rho=0, c=1):
  return 1 / (1 + np.exp(-(rho + c*x)))
```

## 1.2
Voy fijar los valores de $c$ en
*   $c=0$
*   $c>0$
*   $c<0$

y variar el valor de $\rho$ para hacer el análisis gráfico de la incidencia de los parámetros en la función sigmoidea.






```python
x = np.linspace(-10, 10, 400)
rho = [-5,0, 5]

fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Caso c = 0:
for r in rho:
    y = sigmoidea(x, r, 0)
    ax[0].plot(x, y, label=f'rho={r}')

ax[0].axhline(0, color='black', linewidth=0.5)
ax[0].axvline(0, color='black', linewidth=0.5)
ax[0].set_title('Sigmoidea con  c = 0')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Sigmoidea(x)')
ax[0].legend()

# Caso c > 0:
for c in [1,5]:
  for r in rho:
        y = sigmoidea(x, r,c)
        ax[1].plot(x, y, label=f'c={c},rho={r}')

ax[1].axhline(0, color='black', linewidth=0.5)
ax[1].axvline(0, color='black', linewidth=0.5)
ax[1].set_title('Sigmoidea con  c > 0')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Sigmoidea(x)')
ax[1].legend()

# Caso c < 0:
for c in [-1,-5]:
  for r in rho:
      y = sigmoidea(x, r,c)
      ax[2].plot(x, y, label=f'c={c},rho={r}')

ax[2].axhline(0, color='black', linewidth=0.5)
ax[2].axvline(0, color='black', linewidth=0.5)
ax[2].set_title('Sigmoidea con c < 0')
ax[2].set_xlabel('x')
ax[2].set_ylabel('Sigmoidea(x)')
ax[2].legend()

plt.tight_layout()
plt.show()
```

A partir de los gráficos se puede concluir lo siguiente:

*   $\rho$ traslada la curva:

Para $c>0$, $\rho$ traslada la pendiente de la curva, dependiendo de su signo, a lo largo del eje $x$ (hacia la derecha cuando $\rho<0$ y a la izquiera cuando $\rho>0$). Para $c<0$, es análogo al caso anterior, pero invertido. Finalamente, para $c=$0, la función es constante y $\rho$ la traslada hacia arriba mientras más positivo sea su valor y hacia abajo en el caso contrario.
*  $c$ determina la inclinación de la pendiente de la curva:

Si $c>0$ la pendiente es más pronunciada y a medida que el valor de $c$ va disminuyendo esta se va aplanando. Cuando llega a $c=0$ se vuelve constante. Luego, a medida que el valor absoluto de $c<0$ va aumentando, la pendiente se vuelve a hacer pronunciada, pero ahora la curva está al revés.

Ya se vió que tengo 3 posibles puntos fijos que dependen del valor del $\rho$ elegido. Un punto fijo es una raíz del campo vector, como ya se mecionó antes, un punto que cumple $\dot{x}=f(x)=0$, en este caso $f(x)=-x+\frac{1}{1+e^{-(\rho+10x)}}$
.

Para encontrar los puntos fijo voy a usar el método de Newton, y para eso calculo la derivada del campo vector.

```python
def campovector(rho, x):
    return -x + 1 / (1 + np.exp(-(rho + 10 * x)))

def der_campovector(rho, x):
  return -1 + 10 * np.exp(-(rho + 10 * x)) / (1 + np.exp(-(rho + 10 * x)))**2
```

```python
def newton(f, Df,rho, x0, max_iter):
    xn = x0
    for n in range(0, max_iter):
        fxn = f(rho,xn)
        if abs(fxn) < 1e-16:
            return xn
        Dfxn = Df(rho,xn)
        if Dfxn == 0:
            return None
        xn = xn - fxn/Dfxn
    return None
```

Para ver ahora la estabilidad de los puntos fijos voy a evaluarlos en la derivada del campo vector:

*   Si $f'(x^*)>0$ entonces el punto es inestable
*   Si $f'(x^*)<0$ entonces el punto es estable

Además voy comparar los resultados con los gráficos. Para calcular la derivada númerica voy a usar diferencias centradas pues da la mejor aproximación del valor de la derivada en un punto.

```python
def derivar(f,rho, x0, h=1e-8):
  return (f(rho,x0 + h) - f(rho,x0 - h))/(2 * h)
```

Creo una función que gráfique los puntos fijos con sus estabilidades correspondientes (agrego el caso de medio estable, es decir cuando $f(x^*)=0$).

```python
def plot_puntos_fijos_estabilidad(puntosfijos, rho):
    x = np.linspace(-1, 2, 500)
    pendientes = [derivar(campovector, rho, x0) for x0 in puntosfijos]

    estabilidad = []
    for p in pendientes:
        if p < 0:
            estabilidad.append(1)  # Estable
        elif p > 0:
            estabilidad.append(0)  # Inestable
        else:
            estabilidad.append(2)  # Medio estable

    f = -x + sigmoidea(x, rho, 10)
    markerfcolors = ['w', 'r', 'g']
    markerecolors = ['r', 'r', 'g']
    xmarkerstyle = ['<', '>']

    plt.plot(x, f, label='Campo vector', color='blue')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)

    for p in range(len(puntosfijos)):
        puntosfijos_sorted = np.sort(puntosfijos)
        if puntosfijos_sorted[p] == puntosfijos_sorted.min():
            x_markers_tmp = [x[0] + (puntosfijos_sorted[p] - x[0]) / 2]
        elif puntosfijos_sorted[p] == puntosfijos_sorted.max():
            x_markers_tmp = [x[-1] - (x[-1] - puntosfijos[p]) / 2]
        else:
            x_markers_tmp = [puntosfijos_sorted[p] - (puntosfijos_sorted[p] - puntosfijos_sorted[p-1]) / 2,
                             puntosfijos_sorted[p] + (puntosfijos_sorted[p+1] - puntosfijos_sorted[p]) / 2]

        plt.plot(puntosfijos[p], 0, marker='o', markerfacecolor=markerfcolors[estabilidad[p]], markeredgecolor=markerecolors[estabilidad[p]])

        for i in range(len(x_markers_tmp)):
            plt.plot(x_markers_tmp[i], 0, c='k', marker=xmarkerstyle[int(campovector(rho, x_markers_tmp[i]) > 0)])

    stable_marker = plt.Line2D([0], [0], marker='o', color='w', label='Inestable', markerfacecolor='w', markeredgecolor='r', markersize=10)
    unstable_marker = plt.Line2D([0], [0], marker='o', color='w', label='Estable', markerfacecolor='r', markeredgecolor='r', markersize=10)
    halfstable_marker = plt.Line2D([0], [0], marker='o', color='w', label='Medio estable', markerfacecolor='g', markeredgecolor='g', markersize=10)

    plt.legend(handles=[stable_marker, unstable_marker, halfstable_marker])
    plt.xlabel('$x$')
    plt.ylabel('$f(x), \dot{x}$')
    plt.show()

```

Por último, voy a hacer un diagrama de bifurcaciones para entender los cambios cualitativos en el sistema. Para eso voy a graficar $x^*$ vs. $\rho$. Planteando $\dot{x}=0$ y despejando en base a $\rho$ obtengo

$\rho=-\ln(-\frac{x-1}{x})-10x$.
Así que el diagrama de bifurcaciones será sobre la curva con puntos $(x,\rho)$

Las bifurcaciones describen los cambios cualitativos en la dinámica del sistema. Más arriba ví que estos ocurrían en $\rho_1\approx-6.8$ y en $\rho_2\approx-3.2$.

En $\rho<\rho_1$ el $x=0$ es el único punto estable.

En $\rho=\rho_1$ el $x=0$ es estable y el otro punto fijo es medio estable.

En $\rho_1<\rho<\rho_2$ el punto fijo cerca de $x=0$ (que se va a empezar a alejar a medida que aumenta el valor de $\rho$) sigue siendo estable y ahora tengo dos puntos fijos, uno inestable en el medio y uno estable cerca de $x=1$.

En $\rho=\rho_2$ el punto fijo que estaba en $x=0$ y el del medio convergen en un único punto medio estable y $x=1$ es estable.

En $\rho>\rho_2$ el $x=1$ es el único punto estable.

Todo esto tiene que quedar reflejado en el diagrama de bifurcaciones.

```python
rho = np.linspace(0, 1, 800)
x = -np.log(-(rho-1)/rho)-10*rho

ch1 = np.max(np.where(rho <= (5-np.sqrt(15))/10 )) #(5-sqrt(15))/10 es el punto x donde ocurren la bifucarción para rho1
ch2 = np.min(np.where(rho >= (5+np.sqrt(15))/10 )) #(5-sqrt(15))/10 es el punto x donde ocurren la bifucarción para rho2

plt.plot(x[:ch1+1], rho[:ch1+1], 'b', label='Estable')
plt.plot(x[ch1:ch2+1], rho[ch1:ch2+1], 'r', label='Inestable')
plt.plot(x[ch2:], rho[ch2:], 'b')

plt.xlabel(r'$\rho$')
plt.ylabel(r'$x$')
plt.title('Diagrama de bifurcación')
plt.grid(True)
plt.legend()
plt.show()
```

El diagrama refleja los cambios cualitativos: antes de $\rho_1$ tengo un único punto estable en 0, entre $\rho_1$ y hasta $\rho_2$ tengos dos puntos estables cerca de $x=0$ y $x=1$, y uno inestable en $0<x<1$. A partir de $\rho_2$ vuelvo a tener un único punto fijo estable en 1.

Se trata de una bifurcación double fold.

## 1.5

Para calcular las soluciones del sistema voy a integrar por el método de Euler, para los siguiente valores de $\rho:[-8,-6.8,-5]$

```python
def odeEuler(rho,ci, h=0.1, tmax=10):
  t = np.arange(0, tmax, h)
  x = np.zeros(len(t))
  x[0] = ci

  for i in range(len(t) - 1):
    x[i+1] = x[i] + h * campovector(rho,x[i])

  return t, x #vector de tiempos y los valores de la variable en cada paso
```

```python
rho = [-8, -6.8, -5]
condiciones_iniciales = np.arange(-2, 2, 0.1)

fig, axs = plt.subplots(2, 3, figsize=(18, 10))

for idx, rho in enumerate(rho):
    for ci in condiciones_iniciales:
        t, N = odeEuler(rho, ci)
        axs[0, idx].plot(t, N)
    axs[0, idx].set_title(f'rho = {rho}')
    axs[0, idx].set_xlabel("t", fontsize=16)
    axs[0, idx].set_ylabel("x(t)", fontsize=16)
    axs[0, idx].axhline(0, color='black', linewidth=0.5)
    axs[0, idx].axvline(0, color='black', linewidth=0.5)

    x = np.linspace(-2, 2, 400)
    f = campovector(rho,x)
    axs[1, idx].plot(x, f, 'r')
    axs[1, idx].set_xlabel("x", fontsize=16)
    axs[1, idx].set_ylabel("$\dot{x}$", fontsize=16)
    axs[1, idx].axhline(0, color='black', linewidth=0.5)
    axs[1, idx].axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```

Vemos que las formas cualitativas de las soluciones son las esperadas para cada $\rho$.

En el caso de $\rho=-8$ hay un único punto fijo estable en $x=0$, así que todas las soluciones lo aproximan sin importar la condición inicial. Las soluciones con condición inicial menor a 0 decrecen hacia el punto fijo, por eso las curvas de estas soluciones son cóncavas. Para las soluciones con condición inicial mayor a cero la situación es más delicada. Para $0<x<\approx{0.5}$, las soluciones crecen rápidamente hacia el punto fijo, por eso las curvas son convexas. Para $0.5<x<1$, las soluciones tienen una pequeña desaceleración hasta que cruzan 0.5, donde vuelven a crecer, que corresponde con las curvas cóncavas que luego se vuelven convexas. Por último, si $x>1$ entonces las soluciones, crecen, decrecen y luego vuelven a crecer hacia el punto fijo, que de nuevo corresponde con las curvas que comienzan convexas, se vuelven cóncavas y vuelven a ser convexas al final. Por último, si la condición inicial es 0 entonces la solución permanece constante.

En el caso de $\rho=-6.8$ existen dos puntos fijos, $x=0$, que es estable, y $x=\frac{5+\sqrt{15}}{10}$ que es medio estable. Al igual que el caso anterior, las soluciones con condición inicial $x<0$ decrecen casi linealmente, por eso las curvas correspondientes son cóncavas. Las soluciones con condiciones inciales $0<x<1$ también se acercan al punto fijo en 0 (de la misma manera que el caso anterior, algunas crecen, y otras decrecen y vuelven a crecer). Sin embargo, ahora las soluciones con condición inicial mayor a 1 se aproximan al segundo punto fijo crecientemente (en el caso anterior estas también iban al 0), lo que corresponde con las curvas convexas que se ven el gráfico. Por último, si las condiciones iniciales son en los puntos fijos entonces las soluciones se mantienen constantes.

En el caso de $\rho=-5$ existen tres puntos fijos, dos estables y uno inestable. Al igual que los dos casos anteriores, si $x<0$ las soluciones se aproximan decrecientemente al punto fijo en 0. Para $0<x<0.25$ crecen rápidamente al cero, lo que corresponde con las curvas convexas; y si $0.25<x<0.5$, entonces decrecen al comienzo, hasta que cruzan 0.25 y comienzan a crecer (curvas cóncavas que se vuelven convexas). Como el punto fijo en $x=0.5$ es inestable, solo las soluciones constantes lo aproximan. Si $x>0.5$, entonces las soluciones se acercan al punto fijo en $x=1$. Si $0.5<x$ y $x$ es menor que la cima de la pequeña parábola que va hacia 1, entonces primero tienen un crecimiento rápido y luego comienzan a decrecer (corresponde con las curvas convexas que luego se vuelven cóncavas); y si $x$ comienza o es mayor que el valor en esa cima, entonces las soluciones decrecen hacia 1 (curvas cóncavas). Por último, si $x>1$, entonces las soluciones crecen hacia el punto fijo, lo que corresponde con las curvas convexas.

Se pudo ver que en todos los casos las trayectorias de las soluciones corresponden con el comportamiento esperado (determinado por los puntos fijos del sistema).

## 2.4

Voy a definir una función, que dados los puntos fijos de un sistema, me evalúe en la matriz Jacobiana y me devuelva los autovalores correspondientes.

```python
def matriz_jacobiana(x, y):
  dfdx = -1 - 2 * x
  dfdy = -1
  dgdx = 1
  dgdy = -1
  return np.array([[dfdx, dfdy], [dgdx, dgdy]])

def evaluar(puntos_fijos):
  res = []
  for x, y in puntos_fijos:
    j = matriz_jacobiana(x, y)
    eigenvalues, eigenvectors = np.linalg.eig(j)
    res.append(((x, y), j, eigenvalues))
  return res

puntos_fijos = [(-2, -2), (0, 0)]
res = evaluar(puntos_fijos)

for (x, y), j, eigenvalues, in res:
  print(f"Punto fijo: ({x}, {y})")
  print("Matriz jacobiana:")
  print(j)
  print("Autovalores:", eigenvalues)
  print("-" * 20)
```

Con la información obtenida ahora voy a clasificar los puntos fijos:

$(-2,-2)$: En este caso los autovalores son reales y tienen signos opuestos, así que el punto fijo es un punto silla.

$(0,0)$: En este caso los autovalores son complejos conjugados, así que tienen forma de espiral, y además, ambos tienen parte real negativa, así que las soluciones son estables y tienden a acercarse al punto fijo a medida que el tiempo avanza. Por lo tanto, el punto fijo es un foco espiral estable.

Veo primero condiciones iniciales para las trayectorias que se acercan a $(0,0)$.

## 3.1

Defino el sistem SIR y lo resuelvo para ver como evolucionan las poblaciones $S$, $R$ e $I$ en el tiempo

```python
# Modelo SIR
def SIR(t, y, beta, gamma, N):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

N = 10000
beta = 0.2
gamma = 1 / 10
I0 = 1 # una persona enferma
R0 = 0
S0 = N - I0 - R0 # todas las personas que no están enfermas

# Condiciones iniciales
y0 = [S0, I0, R0]

# Intervalo de tiempo
t = np.arange(0, 201, 1)

# Resolver el sistema de ODEs
sol = sp.integrate.solve_ivp(SIR, [0, 200], y0, args=(beta, gamma, N), t_eval=t)

# Graficar las soluciones
plt.figure(figsize=(10, 6))
plt.plot(sol.t, sol.y[0], label='Susceptibles (S)')
plt.plot(sol.t, sol.y[1], label='Infectados (I)')
plt.plot(sol.t, sol.y[2], label='Recuperados (R)')
plt.xlabel('Tiempo (días)')
plt.ylabel('Población')
plt.legend()
plt.title('Evolución del Modelo SIR')
plt.grid(True)
plt.show()
```

Al comienzo empiezan a aumentar los contagios y por lo tanto el número de recuperados. Además, comienza a disminuir las personas susceptibles ya que se empiezan a contagiar. Alrededor del día 90 parecería que hay un pico en la cantidad de contagiados, y se ve que los recuperados siguen aumentando y los susceptibles disminuyendo. Luego de unos días, apox. a partir del día 150, se ve que las poblaciones se vuelven constantes, ya que se dejan de tener contagios.
